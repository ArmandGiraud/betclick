### 1) Error Analysis:

- What is the actual model accuracy on unseen customers?

```python
import seaborn as sns
import matplotlib.pylab as plt
```

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

import sys
sys.path.insert(1, "../src/")

from loader import load_data
from preprocessing import ChrunPrep
data = load_data("../src/data/betclic_datascience_test_churn.csv", 4000000, all_=False)

cp = ChrunPrep()
X = cp.fit_transform(data)

y = cp.create_labels(data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, ExtraTreesClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
lr = LogisticRegression(C=1e10, solver="liblinear", class_weight="balanced", max_iter=1e5)

gb = GradientBoostingClassifier(criterion='friedman_mse', init=None,
                           learning_rate=0.026324967240336495, loss='deviance',
                           max_depth=None, max_features=0.5135820722988753,
                           max_leaf_nodes=None, min_impurity_decrease=0.0,
                           min_impurity_split=None, min_samples_leaf=35,
                           min_samples_split=2, min_weight_fraction_leaf=0.0,
                           n_estimators=78, n_iter_no_change=None,
                           presort='auto', random_state=1,
                           subsample=0.7542915097577992, tol=0.0001,
                           validation_fraction=0.1, verbose=0)


rf = RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
                       max_depth=4, max_features=0.4674665741764695,
                       max_leaf_nodes=None, min_impurity_decrease=0.0,
                       min_impurity_split=None, min_samples_leaf=3,
                       min_samples_split=2, min_weight_fraction_leaf=0.0,
                       n_estimators=341, n_jobs=-1, oob_score=False,
                       random_state=4, verbose=False, warm_start=False)

etf = ExtraTreesClassifier(bootstrap=True, class_weight=None, criterion='entropy',
                     max_depth=None, max_features=0.9159744542919692,
                     max_leaf_nodes=None, min_impurity_decrease=0.0,
                     min_impurity_split=None, min_samples_leaf=4,
                     min_samples_split=2, min_weight_fraction_leaf=0.0,
                     n_estimators=363, n_jobs=-1, oob_score=False, random_state=0,
                     verbose=False, warm_start=False)
```

```python
from sklearn.metrics import accuracy_score, f1_score
```

```python
etf.fit(X_train, y_train)
y_pred = etf.predict(X_test)
y_prob_etf = etf.predict_proba(X_test)
accuracy_score(y_test, y_pred)
```

```python
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)
accuracy_score(y_test, y_pred)
```

```python
gb.fit(X_train, y_train)
y_pred = gb.predict(X_test)
y_prob_gb = gb.predict_proba(X_test)
accuracy_score(y_test, y_pred)
```

```python
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
y_prob_lr = lr.predict_proba(X_test)
accuracy_score(y_test, y_pred_lr)
```

```python
# plot distribution of probabilities:

sns.distplot(y_prob_rf[:,1], label="rf")
sns.distplot(y_prob_gb[:,1], label="gb")
sns.distplot(y_prob_etf[:,1], label="lr");
plt.legend();
```

A bimodal distribution show that the model can distnguish two classes

```python
# averaging probas could give a boost if the models leverage dfferent features
import numpy as np
```

```python
averaged = np.mean([y_prob_gb[:,1], y_prob_rf[:,1], y_prob_etf[:,1]], axis=-0)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
accuracy_score(y_test, averaged>0.5)
```

with the Logistic Regression above 77% of churner are correctly predicted found. 
the accuracy ratio just count the % of correctly predicted curtomers, it is also interesting to check different types of errors.

- **Beyond accuracy**

```python
import scikitplot as skplt

skplt.metrics.plot_confusion_matrix(y_test, y_pred, normalize=False);
```

20% of churners are incorrectly labeled as churners by the model on the test sample. (False Positivee)

while 40% of non churners are misclassifed as churners. (False Negative)


If missing a churner is a more serious issue than incorrectly targeting a non churner the model seems inline with business objective. Otherwise a different threshold could be used to calibrate the model. 

```python
# using a threshold of 0.46 yield less False negative
accuracy_score(y_test, y_prob_gb[:, 1] > 0.46)
```

it is possible to associate an error cost matrix to the confusion matrix in order to optimize the model to business objective.


```python
skplt.metrics.plot_confusion_matrix(y_test, y_prob_gb[:, 1]>0.46, normalize=True);
```

With a campaign based on a 0.46 threshold we can reach around **0.89%** of churners,
while contacting incorrectly 20% of non-churners. 

those thresholds can be found along the precision recall curve: 

```python
skplt.metrics.plot_precision_recall_curve(y_test, y_prob_gb);
```

The RF and GB estimators seems to leverage similar variables(user time (loyalty) and transaction number, while the logistic regression relies on transactions statistics.

- the variance of bet number, seems very important but this varaible does not have a lot of variance.
- the customer segmentation seems not very important and are likely built from the demographic variables already included in the model. 
- Globally it seems that variables that indicates the user activity are the most important, this is a sign of the model health.

### 3) Random Thoughts

1. it is important to note that the model is **undertuned** and can be further improved for example
    - further hyperparmeter selection
    - ensembling/stacking of different model types
    - deeper models selection:
        - in particular the categorical variables seems to have a strong predictive power it seems interesting to try
            an architecture similar to [wide and deep](https://ai.googleblog.com/2016/06/wide-deep-learning-better-together-with.html)
        - categorical variables represented as continuous embeddings could give a great boost when the data is sufficiently large. 
        - in the Exploration notebook it seems that transactions sequences could be leveraged for some users
            RNNs or LSTM could be useful to extract meaningful information from transaction sequences. 
    - Creating/exploring more features
    - trying autoML/automatic feature engeneerings/ features interaction
    - [XgBoost](https://xgboost.readthedocs.io/en/latest/) is known to yield good rersults on churn prediction, but it is necessary to use Docker images for os/plateform compatibility
    
- A massive of **execution time optimizations** can be performed on the current code if training-serving time is a constraint.

    The pandas computations could be optimized by reducing the algorithm complexity, alterntively using a parallel computing framework like Spark/dask can be very efficient.
- it might also be important for **maintenability**
    - write unit tests/e2e tests for the developped functions.
    - Monitor the values of the features on the serving environment, one could use [MLflow](https://mlflow.org/) or plugging a dashboard (after/before the preprocessing step) to detect anomalies or new categories on the test set.
    - Not using a Python application but a scheduler like Apache Airflow for monitoring batch processing,
    - model  versioning ... 
    - use proper python logging instead of print.

It is also advisable to perform A/B testing if the model is used to performed targeted campains. to Validate/maesure the impact of the model. 

Maybe choosing another target would benefitial:
    how to distinguish between a low frequency user (less than every two month) and a churner?
